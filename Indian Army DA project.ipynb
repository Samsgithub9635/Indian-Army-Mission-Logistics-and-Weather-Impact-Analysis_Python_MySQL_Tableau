# Identify Missing Values

```python
# See total and percentage of missing values
def missing_data_summary(df):
    total = df.isnull().sum()
    percent = (total / len(df)) * 100
    return pd.DataFrame({'Missing Values': total, 'Percent': percent})

print(missing_data_summary(logistics))
print(missing_data_summary(weather))
print(missing_data_summary(drone))
```

```python
# Fill missing categorical values with mode (most frequent category)
logistics['Delay_Reason'] = logistics['Delay_Reason'].fillna(logistics['Delay_Reason'].mode()[0])

```

# Detect and Handle Outliers

```python
# Set plot style
sns.set(style="whitegrid")

# Plot outliers in 'Distance_km'
plt.figure(figsize=(8, 4))
sns.boxplot(x=drone['Distance_km'])
plt.title("Outliers in Distance_km")
plt.show()

# Plot outliers in 'Battery_Used_%'
plt.figure(figsize=(8, 4))
sns.boxplot(x=drone['Battery_Used_%'])
plt.title("Outliers in Battery_Used_%")
plt.show()
```

```python
def detect_outliers_all_columns(df, df_name="Logistics"):
    print(f"\n📊 Outlier Detection for {df_name} Dataset:")
    for column in df.select_dtypes(include='number').columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower) | (df[column] > upper)][[column]]
        
        print(f"\n🔍 Column: {column}")
        print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
        print(f"Lower Bound: {lower}, Upper Bound: {upper}")
        print(f"Number of outliers: {len(outliers)}")
        print(outliers.head())  # Show first 5 outliers for preview

```

## Calculate the average delivery speed grouped by each payload size (rounded to nearest 5kg bucket).



```python
query = """
SELECT 
    ROUND(d.Payload_kg / 5) * 5 AS Payload_Bucket,
    AVG(d.Avg_Speed_kmph) AS Avg_Speed
FROM Drone d
GROUP BY Payload_Bucket
ORDER BY Payload_Bucket;
"""

# Execute the query
cur.execute(query)

# Fetch the results
data = cur.fetchall()

# Check and display the results
if not data:
    print("No data found.")
else:
    columns = [desc[0] for desc in cur.description]
    for row in data:
        print(dict(zip(columns, row)))
```

## Find the % of drone deliveries that arrived on time vs delayed.

```python
# Query to calculate on-time and delayed drone deliveries
query = """
SELECT 
    COUNT(*) AS Total_Deliveries,
    SUM(CASE WHEN Status = 'On Time' THEN 1 ELSE 0 END) AS On_Time,
    SUM(CASE WHEN Status = 'Delayed' THEN 1 ELSE 0 END) AS `Delayed`,
    ROUND(SUM(CASE WHEN Status = 'On Time' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS On_Time_Percentage,
    ROUND(SUM(CASE WHEN Status = 'Delayed' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS Delayed_Percentage
FROM Logistics;
"""




# Execute the query
cur.execute(query)

# Fetch the results
data = cur.fetchall()

# Get column names
columns = [desc[0] for desc in cur.description]

# Print the result as dictionary
for row in data:
    print(dict(zip(columns, row)))
```

## Determine the correlation between battery usage and distance (using GROUP BY distance buckets).

```python
query = """
SELECT Distance_km, `Battery_Used_%`
FROM Drone
WHERE Distance_km IS NOT NULL AND `Battery_Used_%` IS NOT NULL;
"""

cur.execute(query)
data = cur.fetchall()

# --- Create DataFrame ---
df = pd.DataFrame(data, columns=["Distance_km", "Battery_Used_%"])

# --- Calculate correlation ---
correlation = df["Distance_km"].corr(df["Battery_Used_%"])
print(f"Correlation between Distance (km) and Battery Used (%): {correlation:.2f}")




```

## Find the most frequent source base–destination post pair and how many times it was used.

```python
query = """
    SELECT 
        Source_Base, 
        Destination_Post, 
        COUNT(*) AS Frequency
    FROM Logistics
    GROUP BY Source_Base, Destination_Post
    ORDER BY Frequency DESC
    LIMIT 1;
"""

# Execute the query
cur.execute(query)

# Fetch the result
result = cur.fetchone()

# Print the result
if result:
    source_base, destination_post, frequency = result
    print(f"The most frequent source base–destination post pair is from '{source_base}' to '{destination_post}', used {frequency} times.")
else:
    print("No data found.")
```

## Identify patterns where snowfall exceeded 10cm and missions were either delayed or failed — highlight reason.

```python
query = """
SELECT 
    l.Mission_ID,
    l.Source_Base,
    l.Destination_Post,
    l.Transport_Mode,
    l.Status,
    w.Date,
    w.Location,
    w.Temperature_C,
    w.Snowfall_cm,
    w.Visibility_m,
    w.Wind_Speed_kmph
FROM Logistics l
JOIN Weather w 
    ON l.Source_Base = w.Location
WHERE 
    (l.Status = 'Delayed' OR l.Status = 'Failed')
    AND w.Snowfall_cm > 10;
"""

cur.execute(query)
data = cur.fetchall()

# Print formatted results
print("Missions affected by heavy snowfall (>10cm):")
print("-" * 80)

for row in data:
    mission_id, source, destination, mode, status, date, location, temp, snow, vis, wind = row
    
    reason = []
    if snow > 10:
        reason.append("Heavy Snowfall")
    if vis < 500:
        reason.append("Low Visibility")
    if wind > 50:
        reason.append("High Wind Speed")

    print(f"""
Mission ID     : {mission_id}
Source         : {source}
Destination    : {destination}
Transport Mode : {mode}
Status         : {status}
Date           : {date}
Snowfall       : {snow} cm
Visibility     : {vis} m
Wind Speed     : {wind} km/h
Possible Reason: {', '.join(reason) if reason else 'Unknown'}
""")
    print("-" * 80)
```
